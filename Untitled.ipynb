## Loading the file into Pandas

```python
import pandas as pd
from pathlib import Path

current_dir = Path.cwd()

def load_personal_data(file_path=current_dir):
    csv_path = current_dir.joinpath("insurance.csv")
    return pd.read_csv(csv_path)
```
