```python
insurance_data = train_set.drop("charges", axis=1)
insurance_labels = train_set["charges"].copy()
```

No missing values in any of the rows, therefore we can continue on to the next process which is to removing text attributes.

```python
smoker_data = insurance_data[["smoker"]]
region_data = insurance_data[["region"]]
gender_data = insurance_data[["sex"]]

from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
smoker_cat_1hot = cat_encoder.fit_transform(smoker_data)
region_cat_1hot = cat_encoder.fit_transform(region_data)
gender_cat_1hot = cat_encoder.fit_transform(gender_data)
```

```python
insurance_num = insurance_data.drop("smoker", axis=1)
insurance_num = insurance_num.drop("region", axis=1)
insurance_num = insurance_num.drop("sex", axis=1)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

num_attribs = list(insurance_num)

smoker_attribs = ["smoker"]
region_attribs = ["region"]
sex_attribs = ["sex"]


num_pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=3)),
    ('std_scaler', StandardScaler()),
    ('model', Ridge())
])

full_pipline = ColumnTransformer([
    ("num_attribs", num_pipeline, num_attribs),
    ("smoker_cat", OneHotEncoder(), smoker_attribs),
    ("region_cat", OneHotEncoder(), region_attribs),
    ("sex_cat", OneHotEncoder(), sex_attribs),
])

insurance_prepared = num_pipeline.fit_transform(insurance_data)
```

```python
print(insurance_prepared)
```

```python
from sklearn.linear_model import LinearRegression


lin_reg = LinearRegression()
lin_reg.fit(insurance_prepared, insurance_labels)
```

```python
some_data = insurance_with_id.iloc[:5]
some_labels = insurance_labels.iloc[:5]

some_data_prepared = full_pipline.transform(some_data)
print("Predictions: " , lin_reg.predict(some_data_prepared))
print(some_labels)
```

```python
from sklearn.metrics import mean_squared_error

insurance_prediction = lin_reg.predict(insurance_prepared)

lin_mse = mean_squared_error(insurance_labels, insurance_prediction)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```
